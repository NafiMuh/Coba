# Feature Selection (Determine Input Features and Target Label) & Preprocessing Data (Data Normalization, etc)

```python
import seaborn as sb

plt.figure(figsize=(8,8))
sb.heatmap(df.corr(), cmap="coolwarm", annot=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Summary'] = label_encoder.fit_transform(df['Summary'])   
df['Summary'].unique()
```

```python
df['Precip_Type'] = label_encoder.fit_transform(df['Precip_Type'])   
df['Precip_Type'].unique()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df['Wind_Speed_kmh'] = scaler.fit_transform(df['Wind_Speed_kmh'].values.reshape(-1,1))   
df['Wind_Speed_kmh']
```

```python
df['Pressure_millibars'] = scaler.fit_transform(df['Pressure_millibars'].values.reshape(-1,1))   
df['Pressure_millibars']
```

```python
df
```

# Splitting Dataset (RatioTrain Set 70% and Ratio Test Set 30%)

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns='Apparent_Temperature_C').values
y = df.Apparent_Temperature_C.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Machine Learning Model Training & Machine Learning Model Testing

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
fx = regressor.predict(X_train)
fx
```

```python
plt.plot(X_train, y_train, color='red')
plt.title('train')
plt.show()
```

```python
plt.plot(X_test, y_test, color='red')
plt.title('test')
plt.show()
```

# Performance Evaluation

```python
print('intersept:', regressor.intercept_)
print('slope:', regressor.coef_[0])
```

```python
r_square = regressor.score(X_train, y_train)
print('coefficient of determination (R^2) for Train Set:', r_square)
```

```python
r_square = regressor.score(X_test, y_test)
print('coefficient of determination (R^2) for Train Set:', r_square)
```
